```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
xfunction = np.array([])

for i in range(n):
    x1 = np.random.uniform(-1, 1)
    x2 = np.random.uniform(0, 1.5)

    while (x2 >= 1.5 * x1 ** 2):
        x1 = np.random.uniform(-1, 1)
        x2 = np.random.uniform(0, 1.5)
    xfunction = np.append(xfunction, x1)

X=np.arange(-1, 1, 0.01)
fx = 1.5 * X ** 2
plt.plot(X, fx)
plt.hist(xfunction, bins=100, density=True)
plt.show()
```
